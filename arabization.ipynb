```python
# StopWords and Punct Removing  Function.
def stop_words(s):
    filtered_text = []
    pn = string.punctuation
    tokens = Tokenize(s)
    stopword_list = stopwords.words('arabic')

    for i in tokens:
        if i not in stopword_list:
            if i not in pn:
                filtered_text.append(i)
    print(filtered_text)
    return filtered_text
```

```python
# Light __ Stemming.
def light_stemmer(s):
    prefix = ""
    suffix = ""
    stem = ""

    tokens=stop_words(s)
    for w in tokens:
        for i in DEFAULT_PREFIX_LIST:
                if w.startswith(i):
                    prefix = i

        for e in DEFAULT_SUFFIX_LIST:
                if w.endswith(e):
                    suffix = e

        stem=w.strip(prefix)
        stem=stem.strip(suffix)
        print (" prefix: "+prefix+" stem: "+stem+" suffix: "+suffix)
        prefix=""
        suffix=""
```

```python
# Root __ Stemming
def root_stemmer(s):
    tokens = stop_words(s)
    st = nltk.ISRIStemmer()
    resultstem = [st.stem(w) for w in tokens]
    for w in resultstem:
        print(w)
```

```python
file1['label'] = file1.label.map({'neg' : 0, 'pos' : 1})
file1['Tweets'] = file1.Tweets.str.replace('[^\w\s]', '')
file1['Tweets'] = file1['Tweets'].apply(nltk.word_tokenize)
file1
```

```python
file1['Tweets'] = file1['Tweets'].apply(lambda x: ' '.join(x))
```

```python
#Split_Dataset to (Train&Test)  
features=vec.transform(file1['Tweets'])
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(features,file1['label'],test_size=0.2)        #train = 80 %  , test = 20 %
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
#Classification using naive_bayes.
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
```

```python
predicted=model.predict(x_test)
#print(len(predicted))
print(np.mean(predicted==y_test))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predicted)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predicted))
```

```python
# GUI _ App
def Gui():
    def btnclear():
        txt.set("")

    root = Tk()
    root.geometry("400x500")
    root.title("Gui_Task")

    txt = StringVar()

    en = Entry(root, textvariable=txt, font=('arial', 16, 'bold'), width=50)
    en.pack()

    # Button 1    ---Diacritics
    btn1 = Button(root, text="Diacritics", border=3, fg="Red",relief=RIDGE ,command=lambda: Diacritics(en.get()))
    btn1.place(x=50, y=100, height=50, width=100)

    # Button 2    ---Tokenize
    btn2 = Button(root, text="Tokenize", border=3, fg="Blue",relief=RIDGE ,command=lambda: Tokenize(en.get()))
    btn2.place(x=200, y=100, height=50, width=100)

    # Button 3    ---Segmentation
    btn3 = Button(root, text="Segmentation", border=3, fg="Green",relief=RIDGE ,command=lambda: segment(en.get()))
    btn3.place(x=50, y=200, height=50, width=100)

    # Button 4    ---StopWords
    btn4 = Button(root, text="StopWords", border=3, fg="Red",relief=RIDGE ,command=lambda: stop_words(en.get()))
    btn4.place(x=200, y=200, height=50, width=100)

    # Button 5    ---Light_stemming
    btn5 = Button(root, text="Light_stem", border=3, fg="Blue",relief=RIDGE ,command=lambda: light_stemmer(en.get()))
    btn5.place(x=50, y=300, height=50, width=100)

    # Button 6    ---Root_stemming
    btn6 = Button(root, text="Root_stem", border=3, fg="Green",relief=RIDGE ,command=lambda: root_stemmer(en.get()))
    btn6.place(x=200, y=300, height=50, width=100)

    # Button 7     ---Clear Order
    btclear = Button(root, text="Clear", border=3, fg="Purple",relief=RIDGE)
    btclear.place(x=150, y=400, height=50, width=100)
    btclear['command'] = btnclear

    root.mainloop()

```
