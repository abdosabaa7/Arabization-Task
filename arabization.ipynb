```python
# StopWords and Punct Removing  Function.
def stop_words(s):
    filtered_text = []
    pn = string.punctuation
    tokens = Tokenize(s)
    stopword_list = stopwords.words('arabic')

    for i in tokens:
        if i not in stopword_list:
            if i not in pn:
                filtered_text.append(i)
    print(filtered_text)
    return filtered_text
```

```python
# Light __ Stemming.
def light_stemmer(s):
    prefix = ""
    suffix = ""
    stem = ""

    tokens=stop_words(s)
    for w in tokens:
        for i in DEFAULT_PREFIX_LIST:
                if w.startswith(i):
                    prefix = i

        for e in DEFAULT_SUFFIX_LIST:
                if w.endswith(e):
                    suffix = e

        stem=w.strip(prefix)
        stem=stem.strip(suffix)
        print (" prefix: "+prefix+" stem: "+stem+" suffix: "+suffix)
        prefix=""
        suffix=""
```

```python
# Root __ Stemming
def root_stemmer(s):
    tokens = stop_words(s)
    st = nltk.ISRIStemmer()
    resultstem = [st.stem(w) for w in tokens]
    for w in resultstem:
        print(w)
```
